# Uniform/ Rectangular Distribution

It is a probability disributions with approximately equally likely outcomes (in rela-world dataset) within a given range.

The uniform distribution encompasses both **discrete and continuous** variants.

### Continuous Uniform Distribution

- In a continuous uniform distribution, outcomes are continuous and infinite.
- **Example:** Heights range 60 inches (5feet) to 72 inches (6 feet).

### Probability Density Function(PDF):

The probability density function of a continuous uniform distribution is constant within a specified interval and zero elsewhere.

where,

a and b are the lower and upper bounds of the interval, respectively.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set the lower and upper bounds of the continuous uniform distribution
lower_bound = 5 #lower bound in inches
upper_bound = 6 #Upper bound in inches 

# Generate random ages from the continuous uniform distribution 
num_people= 1000
ages= np.random.uniform(lower_bound, upper_bound, num_people)
```

```python
# Plot the histogram of ages
plt.figure(figsize=(8,6))
plt.hist(ages, bins=30, density= True, color= 'orange', alpha= 0.7)
plt.title("Histogram of Ages from continuous Uniform Distribution")
plt.xlabel("Age (in inches)")
plt.ylabel("Density")
plt.show()
```

### Characterstics:

- Each value within the interval has an equal probability of occuranc.
- The PDF is a flat line, indicating a constant probability density within the interval.
- The cummulative distribution function (CDF) is a linear function, increasing uniformly from 0 to 1 within the interval.

### Can we plot QQ-plot for Uniform Distribution?

Yes, a QQ plot is a scatterplot created by plotting two sets of quantiles aganist one another.

If both sets of quantiles came from the same distribution, we should see the points forming a line that's roughly straight.

Let's implement it practically.

```python
import scipy.stats as stats
import statsmodels.api as sm
```

```python
# Fit a uniform distribution to the data
params= stats.uniform.fit(ages)
distance= stats.uniform(loc=params[0], scale= params[1])
```

```python
# Create a QQ plot of the data using the uniform distribution
fig= sm.qqplot(ages, dist= distance , line= '45')

# Add a title and lables to the plot
plt.title("QQ Plot of Uniform Distribution with Uniform Fit")
plt.xlabel("Theorietical Quantiles")
plt.ylabel("Sample Quantiles")
plt.show()
```

### Application of Uniform Distribution in Ml and Data Science:

- **Random Initialization:** It is often used for randomly initializing the weights and biases of neural networks and other machine learning models, ensuring randomness and preventing biases during training.
- **Random Sampling:** It is utilized for random sampling techniques like random sampling without replacement and random shuffling of data during model training and evaluation.
- **Hyperparameter Tuning:** It is employed for sampling techniques like random sampling without replacement and random shuffling of data during model grid search or random search algorithms in hyperparameter tuning, helping in optimizing machine learning models.
- **Monte Carlo Simulation:** Uniform distribution serves as the basis for generating random numbers within a specified range in Monte Carlo simulations, which are used for modelling uncertainty and randomness in various scenarios such as finance, engineering and risk analysis.
- **Random Forests:** In decision tree-based algorithm like Random Forests, uniform distribution is used for feature selection and splitting nodes, ensuring unbiasedness and randomness in the tree-building process.
- **Data Augmentation:** It can be used for generating random transformations such as rotations, translations and scaling in data augmentation techniques for improving the diversity and size of training datasets.
