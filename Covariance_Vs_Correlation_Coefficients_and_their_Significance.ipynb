# Covariance Vs Correlation Coefficients and their Significance

## Covariance

It is a statistical measure that describes the degree to which two numerical variables are linearly related.

## Why Covariance, if we have Variance?

There, we discussed about Variance, a statistical measure which is not the true representation of spread. 

It provides insights into the variability of individual variables but does not capture the relationship between multiple variables.

In other words, it doesnt quantifies the extent to which changes in one variables are associated with changes in another variables.

## What problems does Covariance solve?

Covariance helps asses the nature and strength of the relationship between two variables. It indicates whether changes in one variable are positively or negatively associated with changes in another variable.

Covariance is a key input in predictive modelling and regression analysis. It helps modellers understands the interdependence between predictor variable, enabling the development of accurate predictive models.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create a DataFrame with two columns "x" and "y"
data= {"x": [12,45,68, 42, 103], "y": [11, 29, 58, 121, 100]}

data= pd.DataFrame(data)
```

```python
# Caluclate the Covariance between columns "x" and "y"

covariance= data["x"].cov(data["y"])

print("Covariance between 'x' and 'y': ", covariance)
```

```python
#create subplots with only one axis
fig,ax1= plt.subplots(1, 1, figsize= (10, 3))

#plot scatterplot on ax1
ax1.scatter(data["x"], data["y"])

#Set title with covariance value
ax1.set_title("Covariance- "+ str(np.cov(data["x"], data["y"])[0,1]))

#show the plot
plt.show()
```

## Limitation of Covariance

It does not tell us about the strength of the relationship between two variables, causing the magnitude of covariance is affected by the scale of the variables.

```python
#original data covariance 
cov_original= np.cov(data["x"], data["y"])[0,1]
print("Covariance of original data:", cov_original)

"""Here, [0,1] refers to the elements at the first row (index 0) and the second column (index 1) of the covariance matrix,
which corresponding to the covariance between variable x and variable y."""

#Data where each variable is multiplied by 2 covariance
cov_scaled= np.cov(data["x"]* 2, data["y"]*2)[0,1]
print("Covariance of scaled data (each variable multiplied by 2):", cov_scaled)

#Comment explaing the difference
print("\nWhen each variable is multiplied by 2, the covariance is also scaled by a factor of 4.")
```

```python
fig, ax= plt.subplots(1, 3, figsize=(15, 3))

#plot scatter on each axes

#Scatter of variable x aganist iteself
ax[0].scatter(data["x"], data["x"])

#Scatter of variable x aganist variable y
ax[1].scatter(data["x"], data["y"])

#Scatter of variable x multiplied by 2 aganist variable y multiplied by 2
ax[2].scatter(data["x"]*2, data["y"]*2)

ax[0].set_title("Covariance - "+ str(np.cov(data["x"], data["y"])[0,1]))
ax[1].set_title("Covariance - "+ str(np.cov(data["x"], data["y"])[0,1]))
ax[2].set_title("Covariance - "+ str(np.cov(data["x"]*2, data["y"])[0,1]*2))

#show the plot
plt.show()
```

It is fairly proved that by just multiplying with 2, the covariance has increased fourflod.

This not appreciated, as Covariance is not normalized which means it does not have a bounded range. Therefore, it is not suitable for comparing the strength of relationships accross different datasets or variables.

## Disadvantages

1. **Directionality:** Covariance only measures the direction of the relationship between variables (positive or negative), but it does not quantify the strength of the relationship. As a result, covariance alone may not provide sufficient information to access the degree of association between variable accurately. Since the magnitude of covariance is affected by the scale of the variables.
2. **Outliers:** Covariance is sensitive to outliers in the data. Extreme values can disproportionately infulence the covarience caluclation, leading to biased estimates of the relationship between variables. This can result in misleading conclusions about the strength and direction of the relationship.
3. **Independence Assumption:** Covariance assumes that variables are linearly related and that their relationship is stationary over time. In practice, variable may exhibit non-linear relationships or time-varying dependencies, which covariance may not capture accurately.
4. **Not Normalized:** Covariance is not normalized and does not have a bounded range. Therefore, it is not suitable for comparing the strength of relationships accross different datasets or variables. Normalized measures such as correlation coefficients are often preferred for this purpose.

```python
#create subplots with two axes
fig, (ax1, ax2)= plt.subplots(1, 2, figsize= (10, 3))

#plot scatterplots on each axes
ax1.scatter(data["x"], data["x"]) # scatterplot of variable "x" aganist itself
ax2.scatter(data["x"], data["y"]) # scatterplot of variable "x" aganist variable "y"

#set title with correlation coefficients
ax1.set_title("correlation - " + str(data["x"].corr(data["x"])))
ax2.set_title("correlation - " + str(data["x"].corr(data["y"])))
```

1. The correlation of a variable with itself is always 1, as it represents the perfect positive linear relationship.
2. The correlation between variable "x" and "y" represents the strength and direction of their linear relationship.

## How to measure Non-Linear Correlation among variables?

The pearson correlation coefficient, which is the most commonly used measure of correlation, quantifies the strength and direction of the linear relationship between two variables.

However, it is important to note that the Pearson correlation coefficient may still capture some aspects of non-linear relationships, especially if the non-linear is mild.

If the relationship between variables is non-linear, alternative measures of correlation may be more appropiate such as:

1. **Spearman's Rank Correlation:** Spearman's correlation coefficient assesses the strength and direction of the monotonic relationship between two variables, regardless of whether the relationship is linear or non-linear. It based on the ranks of the data rather than the actual values and is often when the relationship between variables is non-linear or the data is ordinal.

2. **Kendall's Tau:** It is another non-parametric measure of correlation that assumes the ordinal association between two variables. It is robust to non-linear relationship and outlier and is often used when the data is ordinal or the relationship is non-linear.

3. **Non-Parametric Correlation:** There are also other non-parametric measures of correlation that can capture non-linear relationships, such as the distance correlation or the maximal information coefficient (MIC). These measures do not assume a specific functional form for the relationship between variables and can capture complex non-linear dependencies.

## Example 5:

Let's say you work for an e-commerce company, and your goal is to increase sales by optimizing online advertising startegies,

You have data on daily advertising spending (in dollars) and daily sales revenue (in dollars) for the past year.

## Correlation:

You start by analyzing the correlation between daily advertising spending and daily sales revenue.

After caluclating the correlation, you find a strong positive correlation between the two variables, indicating that days with higher advertising spending tend to have higher sales revenue.

This correlation suggests an associates between advertising spending and sales but does not imply causation.

Causation:

To determine whether advertising spending directly causes an increase in sales revenue, you decide to conduct a controlled experiment.

You randomly selected a group of customers and expose them to an advertising campaign while keeping another group a control group with no exposure to the campaign. After the campaign, you compare the sales revenue between the exposed group and the control group.

If the exposed group shows as significant increase in sales revenue compared to the control group, you may conclude that the advertising campaign caused the increase in sales. This establishes a casual relationship between advertising spending and sales revenue.

However, if there is no significant difference in sales revenue between the exposed group and the control group, you cannot conclude that the advertising campaign caused the increase in sales.

Other factors may be influencing sales, and the correlation observed earlier may be due to confounding variables.


**Correlation is a far more common relationship between 2 variables than causation**
